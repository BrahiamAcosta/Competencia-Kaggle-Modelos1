```python
df = pd.read_csv('train.csv')
df.head(5)#Visualizar los primeros 5 registros
```

```python
#Cantidad de estudiantes por Estrato
df.FAMI_ESTRATOVIVIENDA.value_counts()
```

```python
#Porcentaje de estudiantes con internet
internet_count = df.FAMI_TIENEINTERNET.value_counts()
plt.pie(internet_count, labels=internet_count.index,explode=(0.05,0), autopct='%1.1f%%', startangle=90, colors=['#99ff99', '#ffff99'])
plt.title('La familia del estudiante cuenta con internet')
plt.axis('equal')
plt.show()
```

```python
#Distribución de estudiantes según el valor de la matricula
df.ESTU_VALORMATRICULAUNIVERSIDAD.value_counts().sort_index().plot(kind='bar')
plt.title('Distribución de estudiantes según el valor de la matricula')
plt.xlabel('Valor de la matricula')
plt.ylabel('Cantidad de estudiantes')
plt.show()
```
