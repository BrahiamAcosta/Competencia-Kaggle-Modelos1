#Entrega 2: Preprocesado de los datos

Previo a generar un modelo predictivo, se va a realizar el correpondiente procesamiento de los datos. Sobre el dataframe generado a partir del archivo train.csv se aplicaran el siguiente procesamiento:<br>
<ul>
  <li>Pasar las variables categóricas con respuestas binarias (Si/No) a 1/0 respectivamente</li>
  <li>Convertir las columnas categóricas con múltiples respuestas a on-hot encoding</li>
  <li>Convertir la variable objetivo a numérica</li>
  <li>Convertir la variable estrato de String a numerica</li>
</ul>

Una vez procesadas las variables, se van a completar los registros nulos de cada una de las columnas. La estrategia de imputación será la moda estadística

```python
traindf = pd.read_csv("train.csv")
```

```python
print(traindf.columns)
traindf.head()
```

##Procesamiento de los datos

###Pasar las columnas con respuestas Si/No a 1/0 (FAMI_TIENEINTERNET y ESTU_PAGOMATRICULAPROPIO)

```python
print(f"Valores actuales de FAMI_TIENEINTERNET: {traindf.FAMI_TIENEINTERNET.unique()}")
print(f"Valores actuales de ESTU:PAGOMATRICULAPROPIO: {traindf.ESTU_PAGOMATRICULAPROPIO.unique()}\n")

print("Pasando a 1/0...\n")

traindf.FAMI_TIENEINTERNET = traindf.FAMI_TIENEINTERNET.map({"No": 0, "Si": 1})
traindf.ESTU_PAGOMATRICULAPROPIO = traindf.ESTU_PAGOMATRICULAPROPIO.map({"No": 0, "Si": 1})

print(f"Valores actuales de FAMI_TIENEINTERNET: {traindf.FAMI_TIENEINTERNET.unique()}")
print(f"Valores actuales de ESTU:PAGOMATRICULAPROPIO: {traindf.ESTU_PAGOMATRICULAPROPIO.unique()}")

traindf.head()
```

###One-hot encoding para las demás variables categóricas

```python
#Valor pagado en la matrícula:

matricula_dummies = pd.get_dummies(traindf.ESTU_VALORMATRICULAUNIVERSIDAD, prefix="Paga")
traindf = pd.concat([traindf, matricula_dummies], axis=1)
traindf.drop("ESTU_VALORMATRICULAUNIVERSIDAD", axis=1, inplace=True)

#Horas de trabajo

estudio_dummies = pd.get_dummies(traindf.ESTU_HORASSEMANATRABAJA, prefix="Trabaja")
traindf = pd.concat([traindf, estudio_dummies], axis=1)
traindf.drop("ESTU_HORASSEMANATRABAJA", axis=1, inplace=True)

print("nuevas columnas:")
traindf.columns
```

```python
traindf.head()
```

###Modificar la variable estrato de la vivienda: eliminar el prefijo

```python
traindf.FAMI_ESTRATOVIVIENDA = traindf.FAMI_ESTRATOVIVIENDA.str.replace("Estrato ","")
traindf.head()
```

###Realizar la codificación ordinal para la variable RENDIMIENTO_GLOBAL

```python
traindf.RENDIMIENTO_GLOBAL.unique()
```

```python
rendimiento_dict = {"bajo":1, "medio-bajo":2, "medio-alto":3, "alto":4}
traindf.RENDIMIENTO_GLOBAL = traindf.RENDIMIENTO_GLOBAL.map(rendimiento_dict)
traindf.head()
```

##Imputar los datos faltantes

```python
traindf.isna().sum()
```

###Estrato de la vivienda

```python
estrato_mode=traindf.FAMI_ESTRATOVIVIENDA.mode()[0]
traindf.FAMI_ESTRATOVIVIENDA = traindf.FAMI_ESTRATOVIVIENDA.fillna(estrato_mode)
traindf.FAMI_ESTRATOVIVIENDA.isna().sum()
```

###Internet

```python
internet_mode=traindf.FAMI_TIENEINTERNET.mode()[0]
traindf.FAMI_TIENEINTERNET = traindf.FAMI_TIENEINTERNET.fillna(internet_mode)
traindf.FAMI_TIENEINTERNET.isna().sum()
```

###Educación de padre y madre

```python
dad_education_mode = traindf.FAMI_EDUCACIONPADRE.mode()[0]
mom_education_mode = traindf.FAMI_EDUCACIONMADRE.mode()[0]
print(dad_education_mode)
print(mom_education_mode)
traindf.FAMI_EDUCACIONPADRE=traindf.FAMI_EDUCACIONPADRE.fillna(dad_education_mode)
traindf.FAMI_EDUCACIONMADRE=traindf.FAMI_EDUCACIONMADRE.fillna(mom_education_mode)
print(f"mom education na: {traindf.FAMI_EDUCACIONMADRE.isna().sum()}, dad education na: {traindf.FAMI_EDUCACIONPADRE.isna().sum()}")
```

###Pagó su propia matrícula

```python
selfpay_mode = traindf.ESTU_PAGOMATRICULAPROPIO.mode()[0]
traindf.ESTU_PAGOMATRICULAPROPIO = traindf.ESTU_PAGOMATRICULAPROPIO.fillna(selfpay_mode)
traindf.ESTU_PAGOMATRICULAPROPIO.isna().sum()
```
